# Text Preprocessing for NLP

## Steps

1. **Load Stopwords:** Uses `stopwords.words('english')` to filter common words.
2. **Preprocess Text:**
   - Converts text to lowercase.
   - Removes special characters (`re.sub(r'[^0-9a-z\s]', ' ', text)`).
   - Tokenizes words by whitespace.
   - Removes stopwords.
3. **Test Function:** Runs on `df['News Headline'][0]` and prints:
   - **Original headline**
   - **Processed tokens**

```python


# Ensure NLTK stopwords are available (requires NLTK data download if not already done)
# nltk.download('stopwords')  # (This line can be run once if stopwords corpus is not downloaded)

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Lowercase the text
    text = text.lower()
    # Remove special characters (keep only alphanumeric and whitespace)
    text = re.sub(r'[^0-9a-z\s]', ' ', text)
    # Tokenize on whitespace
    tokens = text.split()
    # Remove stopwords
    tokens = [tok for tok in tokens if tok not in stop_words]
    return tokens

# Example: preprocess the first headline
sample_text = df['News Headline'][0]
print("Original headline:", sample_text)
print("Processed tokens:", preprocess_text(sample_text))
```

```python
# Load spaCy's English language model
nlp = spacy.load('en_core_web_sm')

# Filter rows where the 'News Headline' has at least 10 words
df_filtered = df[df['News Headline'].apply(lambda x: len(str(x).split()) >= 10)]

# Randomly select two headlines from the filtered DataFrame
selected_sentences_all = df_filtered['News Headline'].sample(n=2, random_state=42)

# Select two random sentences from the dataset
selected_sentences = random.sample(selected_sentences_all.tolist(), 2)
print("Selected Sentences:")
print(f"{selected_sentences[0]}\n{selected_sentences[1]}")

# Process and visualize the dependency parses
for sentence in selected_sentences:
    doc = nlp(sentence)
    displacy.render(doc, style='dep', jupyter=True)
```

```python
def plot_statistical_parser(sentences):
    """
    Plots the constituency (statistical) parse trees for a list of two sentences.
    
    Parameters:
    - sentences: list of str, where each string is a sentence to be parsed.
    
    The function uses spaCy with the benepar pipeline to perform statistical parsing.
    It prints the tree in text form and opens a graphical window for each tree.
    """
    
    # Ensure the spaCy model with benepar is loaded (assuming 'nlp' is already defined globally)
    if "benepar" not in nlp.pipe_names:
        nlp.add_pipe("benepar", config={"model": "benepar_en3"})
    
    for i, sentence in enumerate(sentences, 1):
        # Process the sentence with spaCy
        doc = nlp(sentence)
        # For simplicity, assume each input string is one sentence.
        for sent in doc.sents:
            # Get the constituency parse string added by benepar
            parse_str = sent._.parse_string
            # Convert the parse string into an nltk.Tree object
            tree = nltk.Tree.fromstring(parse_str)
            print(f"\nConstituency parse for sentence {i}: {sentence}")
            print("Parse Tree:")
            tree.pretty_print()  # Print the tree in a structured format
            # Graphically display the tree (this may open a new window)
            tree.draw()
            return

plot_statistical_parser(selected_sentences)
```
